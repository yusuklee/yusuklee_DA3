```python
import pandas as pd
df1 = pd.read_csv('(사람인) 개발자 채용 데이터.csv')
df2 = pd.read_csv('(잡코리아) 개발자 채용 데이터.csv')


df =pd.concat((df1,df2), axis=0)
df.head()
```

```python
df.shape
```

```python
tmp = df.copy()
tmp['날짜만'] = tmp['due'].str.extract(r'([0-9]+/[0-9]+)')
tmp.head()
tmp['due'].unique()
```

```python
#날짜만 부분이 nan인친구들은 내일마감이라는 애들이라 오늘날짜 + 1로 하겠다
from datetime import datetime, timedelta

tmp['날짜만']=tmp['날짜만'].fillna(  (datetime.now()+timedelta(days=1)).date().strftime("%m/%d")   )
tmp['날짜만'].unique()
```

```python
tmp = tmp.drop(columns='due')
tmp.head()
```

```python
tmp['title'] = tmp['title'].str.strip()
tmp['company'] = tmp['company'].str.strip()
tmp['link'] = tmp['link'].str.strip()
tmp['loc'] = tmp['loc'].str.strip()
tmp['requirements'] = tmp['requirements'].str.strip()
tmp['날짜만'] = tmp['날짜만'].str.strip()

tmp.head()
```

```python
tmp.rename(columns={'title':'모집 제목', "company":'회사', 'loc':'위치', 'requirements':'필요 조건', '날짜만':'마감일'},inplace=True)
```

```python
tmp.shape
```

```python
tmp = tmp.drop_duplicates(subset=['모집 제목', '마감일'],keep="first")
```

```python
#신입은 이공고중에 몇개 정도 있는지 알고싶음
신입만 = tmp[tmp['필요 조건'].str.contains('신입|경력 무관|경력 무관')]
```

```python
tmp.to_csv('(개발자) 전체 채용 공고.csv', encoding='utf-8-sig')
tmp.to_csv('(개발자) 신입지원 가능한 채용공고.csv')
```
